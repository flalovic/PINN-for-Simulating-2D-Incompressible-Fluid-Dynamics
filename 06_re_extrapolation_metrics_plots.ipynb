```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["font.size"] = 11

df = (
    pd.read_csv("results/re_extrapolation_results.csv")
      .sort_values("c_physics")
)
```

```python
plt.figure()

plt.plot(
    df["c_physics"],
    df["R2_all"],
    marker="o",
    linewidth=2,
)

plt.xlabel(r"$c_{physics}$")
plt.ylabel(r"$R^2$")
plt.title("Overall extrapolation")
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure()

plt.plot(
    df["c_physics"],
    df["RelL2_all"],
    marker="o",
    linewidth=2,
)

plt.xlabel(r"$c_{physics}$")
plt.ylabel("Relative L2")
plt.title("Overall extrapolation")
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
metrics = [
    ("R2_U_x", r"$u_x$"),
    ("R2_U_y", r"$u_y$"),
    ("R2_p", "Pressure"),
    ("R2_all", "Overall"),
]

fig, axs = plt.subplots(2, 2, figsize=(10, 7))
axs = axs.ravel()

for ax, (metric, title) in zip(axs, metrics):

    ax.plot(
        df["c_physics"],
        df[metric],
        marker="o",
        linewidth=2,
    )

    ax.set_title(title)
    ax.set_xlabel(r"$c_{physics}$")
    ax.set_ylabel(r"$R^2$")
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
metrics = [
    ("RelL2_U_x", r"$u_x$"),
    ("RelL2_U_y", r"$u_y$"),
    ("RelL2_p", "Pressure"),
    ("RelL2_all", "Overall"),
]

fig, axs = plt.subplots(2, 2, figsize=(10, 7))
axs = axs.ravel()

for ax, (metric, title) in zip(axs, metrics):

    ax.plot(
        df["c_physics"],
        df[metric],
        marker="o",
        linewidth=2,
    )

    ax.set_title(title)
    ax.set_xlabel(r"$c_{physics}$")
    ax.set_ylabel("Relative L2")
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
heat = (
    df.set_index("c_physics")[
        [
            "R2_U_x",
            "R2_U_y",
            "R2_p",
            "R2_all",
        ]
    ]
    .T
)

heat.index = [
    r"$u_x$",
    r"$u_y$",
    "Pressure",
    "Overall",
]

display(heat)
```

```python
plt.figure(figsize=(8,3))

sns.heatmap(
    heat,
    annot=True,
    fmt=".3f",
    cmap="viridis",
)

plt.xlabel(r"$c_{physics}$")
plt.ylabel("Component")
plt.title(r"$R^2$ on extrapolation")

plt.tight_layout()
plt.show()
```

```python
heat = (
    df.set_index("c_physics")[
        [
            "RelL2_U_x",
            "RelL2_U_y",
            "RelL2_p",
            "RelL2_all",
        ]
    ]
    .T
)

heat.index = [
    r"$u_x$",
    r"$u_y$",
    "Pressure",
    "Overall",
]

display(heat)
```

```python
plt.figure(figsize=(8,3))

sns.heatmap(
    heat,
    annot=True,
    fmt=".3f",
    cmap="viridis_r",
)

plt.xlabel(r"$c_{physics}$")
plt.ylabel("Component")
plt.title("Relative L2 on extrapolation")

plt.tight_layout()
plt.show()
```
