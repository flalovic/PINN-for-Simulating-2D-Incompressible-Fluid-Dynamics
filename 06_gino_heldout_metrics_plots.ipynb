```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["font.size"] = 11

windows = {
    "25%": "results/gino_heldout_results_25.csv",
    "49%": "results/gino_heldout_results_49.csv",
    "64%": "results/gino_heldout_results_64.csv",
}

dfs = {}

for name, path in windows.items():
    dfs[name] = pd.read_csv(path)
```

```python
plt.figure()

for name, df in dfs.items():

    inside = (
        df[df["region"] == "inside_box"]
        .sort_values("c_physics")
    )

    plt.plot(
        inside["c_physics"],
        inside["R2_all"],
        marker="o",
        linewidth=2,
        label=name
    )

plt.xlabel(r"$c_{physics}$")
plt.ylabel(r"$R^2$")
plt.title("GINO — inside removed region")
plt.grid(alpha=0.3)
plt.legend(title="Window size")

plt.tight_layout()
plt.show()
```

```python
plt.figure()

for name, df in dfs.items():

    inside = (
        df[df["region"] == "inside_box"]
        .sort_values("c_physics")
    )

    plt.plot(
        inside["c_physics"],
        inside["RelL2_all"],
        marker="o",
        linewidth=2,
        label=name
    )

plt.xlabel(r"$c_{physics}$")
plt.ylabel("Relative L2")
plt.title("GINO — inside removed region")
plt.grid(alpha=0.3)
plt.legend(title="Window size")

plt.tight_layout()
plt.show()
```

```python
plt.figure()

for name, df in dfs.items():

    outside = (
        df[df["region"] == "outside_box"]
        .sort_values("c_physics")
    )

    plt.plot(
        outside["c_physics"],
        outside["R2_all"],
        marker="o",
        linewidth=2,
        label=name
    )

plt.xlabel(r"$c_{physics}$")
plt.ylabel(r"$R^2$")
plt.title("GINO — outside removed region")
plt.grid(alpha=0.3)
plt.legend(title="Window size")

plt.tight_layout()
plt.show()
```

```python
heat = {}

for window, df in dfs.items():

    values = (
        df[df["region"] == "inside_box"]
        .sort_values("c_physics")
        .set_index("c_physics")["R2_all"]
    )

    heat[window] = values

heat = pd.DataFrame(heat).T

display(heat)
```

```python
plt.figure(figsize=(8,3))

sns.heatmap(
    heat,
    annot=True,
    fmt=".3f",
    cmap="viridis"
)

plt.xlabel(r"$c_{physics}$")
plt.ylabel("Removed window")
plt.title(r"GINO — $R^2$ inside removed region")

plt.tight_layout()
plt.show()
```
